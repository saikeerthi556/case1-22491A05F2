```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
image=cv2.imread("keerthi-suresh.jpg")
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(image)
plt.show()
```

```python
resized=cv2.resize(image,(256,256))
plt.imshow(resized)
plt.show()
```

```python
flip=cv2.flip(image,1)
plt.imshow(flip)
plt.show()
```

```python
blur=cv2.GaussianBlur(image,(55,55),0)
plt.imshow(blur)
plt.show()
```

```python
gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
edges=cv2.Canny(gray,threshold1=50,threshold2=100)
plt.imshow(edges,cmap="gray")
plt.show()
```
